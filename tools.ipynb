Ini tools untuk memecah file besar banyak baris agar  mudah dicopas ke gsheet

```python
df=pd.read_excel("Sept2024_clean.xlsx")

# Tentukan jumlah baris per bagian
baris_per_bagian = 3000

# Bagi data menjadi 3 bagian
data1 = df.iloc[:baris_per_bagian]
data2 = df.iloc[baris_per_bagian:baris_per_bagian*2]
data3 = df.iloc[baris_per_bagian*2:]

# Simpan setiap bagian ke dalam file CSV
data1.to_csv('Data1.csv', index=False)
data2.to_csv('Data2.csv', index=False)
data3.to_csv('Data3.csv', index=False)
```

```python
# Baca data dari file Excel
df1 = pd.read_excel('QUA_Jan_Aug24_FinalCleaningPy3.xlsx')  # Ganti 'file1.xlsx' dengan nama file Excel Anda
df2 = pd.read_excel('Sept2024_clean.xlsx')  # Ganti 'file2.xlsx' dengan nama file Excel Anda

# Gabungkan kedua DataFrame secara vertikal (baris demi baris)
df = pd.concat([df1, df2], ignore_index=True)

# Simpan DataFrame gabungan ke dalam file Excel baru
df.to_excel('df_gabungan.xlsx', index=False)

df
```
